Atlikom kelet eksperimentu: Tikslas sumainti overfitting ir pagerinti out of sample rezultatus strategijai GOPS.

Tam tikslui naudojome:

1 Triukmo pridjim prie finansinio instrumento, strategijos optimizacijos metu. 

2 Bandme optimizuoti strategija ant keliu instrument. 

3 Bandme filtruoti optimizacijos rezultatus neuroninio tinklo pagalba. Neuroninis tinklas turjo perirti optimizacijos 
rezultatus ir atrinkti tuos sprendinius i kuriu butu tikslinga konstruoti portfeli tikintis ateityje geresniu rezultat.

Portfelio konstravimui buvo naudojamas comsha algoritmas, kuris i optimizacijos rezultatu arba tinklo filtravimo metu gautu strategij, atrinkdavo 30 strategij ir i j sukurdavo portfeli.

Visais atvejais buvo naudojamas @ES simbolis. Strategija GOPS, o 30 strategij reikia tai kad i portfeli bdavo dedama 30 GOPS strategij tik su skirtingais parametrais.

Triukmo  pridjimo atveju strategija buvo optimizuojama ant 2 simboli  @ES  ir modifikuoto @ES, buvo modifikuojama OHLC pridedant atsitiktinio triukmo.

Keliu simboli atveju strategija buvo optimizuota ant 3 simboli @ES @NK @NQ.
