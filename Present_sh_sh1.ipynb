Presentacija sh >> sh1

Atlikome keletą eksperimentų: tikslas sumažinti overfitting ir pagerinti out of sample rezultatus strategijai GOPS.

Tam tikslui naudojome:

1 Triukšmo pridėjimą prie finansinio instrumento strategijos optimizacijos metu. 

2 Bandėme optimizuoti strategiją su keliais instrumentais. 

3 Bandėme filtruoti optimizacijos rezultatus neuroninio tinklo pagalba. Neuroninis tinklas turėjo peržiūrėti optimizacijos rezultatus ir atrinkti tuos sprendinius, iš kurių būtų tikslinga konstruoti portfelį tikintis ateityje geresnių rezultatų.

Portfelio konstravimui buvo naudojamas comsha algoritmas, kuris iš optimizacijos rezultatų arba tinklo filtravimo metu gautų strategijų atrinkdavo 30 strategijų ir iš jų sukurdavo portfelį.

Visais atvejais buvo naudojamas @ES simbolis. Strategija GOPS, o 30 strategijų - reiškia tai, kad į portfelį būdavo dedama 30 GOPS strategijų, tik su skirtingais parametrais.

Triukšmo  pridėjimo atveju strategija buvo optimizuojama su 2 simboliais: @ES  ir modifikuotu @ES, buvo modifikuojama OHLC pridedant atsitiktinio triukšmo.

Kelių simbolių atveju strategija buvo optimizuota su 3 simboliais: @ES @NK @NQ.

Neuroninio Tinklo atveju OOS buvo padalinta į 2 dalis ir tinklui buvo leidžiama naudoti IS ir 1 OOS.

OOS periodo pradžia 2017.07.01 

Atitinkamai:

1 OOS pradžia 2017.07.01 

2 OOS pradžia 2020.07.01 

IS pradžia 2003.01.01

Portfelis buvo konstruojamas 2 būdais: naudojant visus IS duomenis ir WF kas 1 mėn.
