Presentacija sh >> sh1

Atlikom keletą eksperimentu: Tikslas sumažinti overfitting ir pagerinti out of sample rezultatus strategijai GOPS.

Tam tikslui naudojome:

1 Triukšmo pridėjimą prie finansinio instrumento, strategijos optimizacijos metu. 

2 Bandėme optimizuoti strategija ant keliu instrumentų. 

3 Bandėme filtruoti optimizacijos rezultatus neuroninio tinklo pagalba. Neuroninis tinklas turėjo peržiūrėti optimizacijos 
rezultatus ir atrinkti tuos sprendinius iš kuriu butu tikslinga konstruoti portfeli tikintis ateityje geresniu rezultatų.

Portfelio konstravimui buvo naudojamas comsha algoritmas, kuris iš optimizacijos rezultatu arba tinklo filtravimo metu gautu strategijų, atrinkdavo 30 strategijų ir iš jų sukurdavo portfeli.

Visais atvejais buvo naudojamas @ES simbolis. Strategija GOPS, o 30 strategijų reiškia tai kad i portfeli būdavo dedama 30 GOPS strategijų tik su skirtingais parametrais.

Triukšmo  pridėjimo atveju strategija buvo optimizuojama ant 2 simbolių  @ES  ir modifikuoto @ES, buvo modifikuojama OHLC pridedant atsitiktinio triukšmo.

Keliu simbolių atveju strategija buvo optimizuota ant 3 simbolių @ES @NK @NQ.

Neuroninio Tinklo atveju OOS buvo padalinta I 2 dalis ir tinklui buvo leidžiama naudoti IS ir 1 OOS.
